```
from pyspark.sql import functions as F 
from pyspark.sql.functions import when,col
```

```
df1 = spark.read.format("csv") \
  .option("inferSchema", "true") \
  .option("header", "true") \
  .option("sep", ",") \
  .load("/FileStore/shared_uploads/Amit.Naha@agilisium.com/city_day.csv")
df2 = spark.read.format("csv").option("header", "true").load("dbfs:/FileStore/tables/station_hour.csv")
```

```
new_cols=(column.replace('.', '_') for column in df1.columns)
df1 = df1.toDF(*new_cols)
display(df1)
```

```
from pyspark.sql.functions import when,col
df_main1 = df1.withColumn("AQI_Levels",when((df1.AQI>=0)&(df1.AQI<=100),"Accepted levels")
                                      .when((df1.AQI>=101)&(df1.AQI<=500),"Unaccepted levels")
                                      .otherwise("Missing values"))


display(df_main1)
```

```
PM10_Top_Contributers = df1.groupBy("City").sum("PM10")
display(PM10_Top_Contributers.dropna().sort(col("sum(PM10)").desc()))
```

```
PM2_5_Top_Contributers = df1.groupBy("City").sum("PM2_5")
display(PM2_5_Top_Contributers.dropna().sort(col("sum(PM2_5)").desc()))
```

```
display(df1.groupBy("City").sum("NO").dropna().sort(col("sum(NO)").desc()).head(5))
```

```
from pyspark.sql.functions import when,col
df_main1 = df1.withColumn("AQI_Quality",when((df1.AQI>=0)&(df1.AQI<=50),"Good")
                                      .when((df1.AQI>=51)&(df1.AQI<=100),"Satisfactory")
                                      .when((df1.AQI>=101)&(df1.AQI<=200),"Moderately Polluted")
                                      .when((df1.AQI>=201)&(df1.AQI<=300),"Poor")
                                      .when((df1.AQI>=301)&(df1.AQI<=400),"Very Poor")
                                      .when((df1.AQI>=401)&(df1.AQI<=500),"Severe")
                                      .otherwise("Bad"))


display(df_main1)
```

```
df_main2 = df_main1.withColumn("AQI_Diseases",when((df_main1.AQI>=0)&(df_main1.AQI<=50),"Minimal Impact")
                                      .when((df_main1.AQI>=51)&(df_main1.AQI<=100),"Minor breathing difficulties")
                                      .when((df_main1.AQI>=101)&(df_main1.AQI<=200),"Asthma , heart disease, children and older adults.")
                                      .when((df_main1.AQI>=201)&(df_main1.AQI<=300),"Breathing difficulties in people on prolonged exposure")
                                      .when((df_main1.AQI>=301)&(df_main1.AQI<=400),"Respiratory illness in people on prolonged exposure")
                                      .when((df_main1.AQI>=401)&(df_main1.AQI<=500),"Respiratory issues in healthy people and Serious lung diseases")
                                      .otherwise("No Diseases"))


display(df_main2)
```

```
agg_cols = ["City","Date"]
df1_grouped = (
    df1.groupBy(agg_cols)
    .agg(
    F.avg(F.col("NO")).alias("Avg_NO"),
    F.avg(F.col("NO2")).alias("Avg_NO2"),
    F.avg(F.col("CO")).alias("Avg_CO"),
    F.avg(F.col("O3")).alias("Avg_O3"),
    )
)
display(df1_grouped)
```
